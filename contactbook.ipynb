```python
from tkinter import *
from tkinter import ttk
from tkinter.ttk import Combobox
from tkinter import messagebox
import sqlite3 as sl

labels = ['Name', 'Mobile No.', 'Telephone No.', 'Email-ID', 'Address', 'Relation']
rel = ['Mother', 'Father', 'Sister', 'Brother', 'Friend', 'Cousin']


#   database creation
try:
    con = sl.connect(database= "contactbookdb.sqlite3")
    cur = con.cursor()
    cur.execute("""CREATE TABLE contactbook(
                cb INTEGER PRIMARY KEY AUTOINCREMENT, 
                img TEXT,
                name TEXT,
                mob INTEGER,
                telephone INTEGER,
                email TEXT,
                address TEXT,
                relation TEXT
                )""")
    con.commit()
    print("Table created successfully.")
except:
    print("Something went wrong! Might be table already created.")
con.close()

win = Tk()
win.geometry("550x700")
win.resizable(width = False, height = False)
win.configure(bg = "thistle1")
win.title("ContactBook")
icon = PhotoImage(file = "human-icon.png")
win.iconphoto(False, icon)


h_frame = Frame(win)
h_frame.configure(bg = "white")
h_frame.place(relx = 0, rely = 0, relwidth = 1, relheight = 0.075)

l_head1 = Label(h_frame, text = "Contact Book", image = icon, bg = "white", fg = "purple")
l_head1.grid(row = 0, column = 0, padx = (100, 0))
l_head = Label(h_frame, text = "Contact Book", font = ("Arial", 24, "bold", "underline"), bg = "white", fg = "purple", width = 15)
l_head.grid(row = 0, column = 1)


def homescreen():
    m_frame = Frame(win)
    m_frame.configure(bg = "thistle1")
    m_frame.place(relx = 0, rely = 0.076, relwidth = 1, relheight = 0.9)

    def contactlist():
       
        sc_frame = Frame(m_frame)
        sc_frame.configure(bg = "thistle1")
        sc_frame.place(relx = 0, rely = 0.001, relheight = 0.93, relwidth = 1)
        
        canvas = Canvas(sc_frame)
        canvas.configure(bg = "thistle1", bd = 0)
        canvas.pack(side = LEFT, fill = 'both', expand = True)
        
        cl_scrollbar = ttk.Scrollbar(sc_frame, orient = "vertical", command = canvas.yview)
        cl_scrollbar.pack(side = RIGHT, fill = "y")

        canvas.configure(yscrollcommand = cl_scrollbar.set)
        canvas.bind('<Configure>', lambda e: canvas.configure(scrollregion=canvas.bbox("all")))

        cl_frame = Frame(canvas)
        cl_frame.configure(bg = "thistle1")
        canvas.create_window((50, 0), window = cl_frame, anchor = "nw")
        
# ***************************************** This is Contact Details Display Screen *****************************************************

        def contactdetails(cb_id):
            m_frame.destroy()

            cd_m_frame = Frame(win)
            cd_m_frame.configure(bg = "thistle1")
            cd_m_frame.place(relx = 0, rely = 0.08, relwidth = 1, relheight = 0.92)

            cd_frame = Frame(cd_m_frame)
            cd_frame.configure(bg = "thistle1")
            cd_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)
            #print(cb_id, type(cb_id))

            # database extraction
            conobj = sl.connect(database= "contactbookdb.sqlite3")
            cur = conobj.cursor()
            cur.execute("SELECT * FROM contactbook  WHERE cb = ?", (str(cb_id)))
            data = cur.fetchone()
            conobj.close()

            #   contact details display
            for j in range(len(labels)):
                cd_label = Label(cd_frame, text = labels[j], width = 12, fg = "black", font = ("arial", 16, ""), bg = "thistle1", anchor ='w')
                cd_label.grid(row = j, column = 0, pady = 10)

                cd_details = Label(cd_frame, text = data[j+2], width = 25, wraplength = 300, font = ("calibri", 16, ""), fg = "purple4", bd = 3, bg = "thistle1", anchor = "w")
                cd_details.grid(row = j, column = 2, pady = 10)

# ************************************************ This Edit/Update Contact Details ***************************************************

            def edit():
                cd_m_frame.destroy()

                edit_m_frame=Frame(win)
                edit_m_frame.configure(bg = "thistle1")
                edit_m_frame.place(relx = 0, rely = 0.1, relwidth = 1, relheight = 0.9)

                edit_frame=Frame(edit_m_frame)
                edit_frame.configure(bg = "thistle1")
                edit_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)

                def back2cd():
                    edit_m_frame.destroy()
                    btn_frame.destroy()
                    contactdetails(cb_id)

                def updatedb():
                    n = e_edit_name.get()
                    m = e_edit_mob.get()
                    t = e_edit_tele.get()
                    e = e_edit_email.get()
                    a = e_edit_addr.get()
                    r = e_edit_rel.get()

                    #   database updation
                    conobj = sl.connect(database = "contactbookdb.sqlite3")
                    cur = conobj.cursor()
                    cur.execute("UPDATE contactbook SET name = ?, mob = ?, telephone = ?, email = ?, address = ?, relation = ? WHERE cb = ?", (n, m, t, e, a, r, cb_id))
                    conobj.commit()
                    conobj.close()
                    messagebox.showinfo("Message", "Contact Updated successfully!")
                    edit_m_frame.destroy()
                    contactdetails(cb_id)
                    
                
                for i in range(len(labels)):
                    l_edit_name = Label(edit_frame, text = labels[i], fg = "black", font = ("arial", 16, ""), bg = "thistle1")
                    l_edit_name.grid(row = i+1, column = 0, pady = 10)

                e_edit_name = Entry(edit_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
                e_edit_name.grid(row = 1, column = 2, pady = 10)

                e_edit_mob = Entry(edit_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
                e_edit_mob.grid(row = 2, column = 2, pady = 10)                

                e_edit_tele = Entry(edit_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
                e_edit_tele.grid(row = 3, column = 2, pady = 10)              

                e_edit_email = Entry(edit_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
                e_edit_email.grid(row = 4, column = 2, pady = 10)

                e_edit_addr = Entry(edit_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
                e_edit_addr.grid(row = 5, column = 2, pady = 10)

                e_edit_rel = Combobox(edit_frame, width = 25, font = ("calibri", 14, "bold"), values = rel)
                e_edit_rel.grid(row = 6, column = 2, pady = 10)

                #print((data))
                e_edit_name.insert(0, data[2])
                e_edit_mob.insert(0, data[3])
                e_edit_tele.insert(0, data[4])
                e_edit_email.insert(0, data[5])
                e_edit_addr.insert(0, data[6])
                r =rel.index(data[7])
                e_edit_rel.current(r)

                btn_frame = Frame(edit_m_frame)
                btn_frame.configure(bg = "thistle1")
                btn_frame.place(relx = 0, rely = 0.93, relwidth = 1, relheight = 0.07)

                b_edit_back = Button(btn_frame, text = "Back", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = back2cd)
                b_edit_back.grid(row = 0, column = 0, padx = 1.5, pady = 0)

                b_edit_save = Button(btn_frame, text = "Save", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = updatedb)
                b_edit_save.grid(row = 0, column = 1, padx = 1.5, pady = 0)
            
            def back2list():
                cd_m_frame.destroy()
                homescreen()


            def deletedb():
                conobj = sl.connect(database= "contactbookdb.sqlite3")
                cur = conobj.cursor()
                cur.execute("DELETE contactbook WHERE cb = ?", (cb_id))
                conobj.commit()
                conobj.close()
                
                cd_m_frame.destroy()
                homescreen()


            btn_frame = Frame(win)
            btn_frame.configure(bg = "thistle1")
            btn_frame.place(relx = 0, rely = 0.9375, relwidth = 1, relheight = 0.07)

            b_cd_back = Button(btn_frame, text = "Delete", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = deletedb)
            b_cd_back.grid(row = 0, column = 0, padx = 1.5, pady = 0)

            b_cd_save = Button(btn_frame, text = "Edit", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = edit)
            b_cd_save.grid(row = 0, column = 1, padx = 1.5, pady = 0)

            b_cd_back = Button(cd_m_frame, text = "Back", width = 5, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = back2list)
            b_cd_back.place(relx = 0, rely = 0)

# ***************************************** This is Contact List Function - Main Screen ************************************************

        #   database extraction to show contacts
        conobj = sl.connect(database= "contactbookdb.sqlite3")
        cur = conobj.cursor()
        cur.execute("select * from contactbook")
        data = cur.fetchall()
        conobj.close()
        #print(data)
        
        for i in range(len(data)):
            c = data[i]
            print(i)
            #text = c[2][0].upper()
            icon = PhotoImage(file = "human-icon.png")
            b_cl_img = Button(cl_frame, image = icon, fg = "black", bg = "thistle1", font = ("arial", 18, "bold"), bd = 0) 
            b_cl_img.image = icon
            b_cl_img.grid(row = i, column = 0, pady = 5)

            b_cl_name = Button(cl_frame, text = c[2], width = 32, font = ("Calibri", 18, ""), fg = "black", bg = "thistle1", bd = 0, command = lambda i = i : contactdetails(i+1))
            b_cl_name.grid(row = i, column = 2, pady = 5)

# ************************************* This is "New Contact" Function to Add New Contact **************************************************

    def newcontact():
        m_frame.destroy()
        nc_m_frame = Frame(win)
        nc_m_frame.configure(bg = "thistle1")
        nc_m_frame.place(relx = 0, rely = 0.1, relwidth = 1, relheight = 0.9)

                
        nc_frame = Frame(nc_m_frame)
        nc_frame.configure(bg = "thistle1")
        nc_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)
        
        l_nc_name = Label(nc_frame, text = "Name:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_name.grid(row = 1, column = 0, pady = 10)

        e_nc_name = Entry(nc_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
        e_nc_name.grid(row = 1, column = 2, pady = 10)

        l_nc_mob = Label(nc_frame, text = "Mobile No.:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_mob.grid(row = 2, column = 0, pady = 10)

        e_nc_mob = Entry(nc_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
        e_nc_mob.grid(row = 2, column = 2, pady = 10)

        l_nc_tele = Label(nc_frame, text = "Telephone No.:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_tele.grid(row = 3, column = 0, pady = 10)

        e_nc_tele = Entry(nc_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
        e_nc_tele.grid(row = 3, column = 2, pady = 10)

        l_nc_email = Label(nc_frame, text = "Email ID:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_email.grid(row = 4, column = 0, pady = 10)         
        
        e_nc_email = Entry(nc_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
        e_nc_email.grid(row = 4, column = 2, pady = 10)

        l_nc_addr = Label(nc_frame, text = "Address:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_addr.grid(row = 5, column = 0, pady = 10)

        e_nc_addr = Entry(nc_frame, width = 25, font = ("calibri", 16, "bold"), fg = "purple", bd = 3)
        e_nc_addr.grid(row = 5, column = 2, pady = 10)

        l_nc_rel = Label(nc_frame, text = "Relation:", fg = "black", font = ("arial", 16, ""), bg = "thistle1")
        l_nc_rel.grid(row = 6, column = 0, pady = 10)

        e_nc_rel = Combobox(nc_frame, width = 23, font = ("calibri", 16, "bold"), values = rel)
        e_nc_rel.current(0)
        e_nc_rel.grid(row = 6, column = 2, pady = 10)

        def newcontactdb():
            name = e_nc_name.get()
            email = e_nc_email.get()
            mob = e_nc_mob.get()
            tele = e_nc_tele.get()
            addr = e_nc_addr.get()
            rel = e_nc_rel.get()

            if name == "" and mob == "":
                messagebox.showwarning(title = "Warning", message = "Invalid input. Enter the valid Details!")
                contactlist()
            else: 
                pass;

            #   database insertion
            conobj = sl.connect(database= "contactbookdb.sqlite3")
            cur = conobj.cursor()
            cur.execute("INSERT INTO contactbook(name, mob, telephone, email, address, relation) VALUES(?, ?, ?, ?, ?, ?)", (name, mob, tele, email, addr, rel))
            conobj.commit()
            conobj.close()
            messagebox.showinfo("Message", "New contact added successfully!")
            nc_m_frame.destroy()
            homescreen()

        def back2home():
            nc_m_frame.destroy()
            homescreen()

        btn_frame = Frame(nc_m_frame)
        btn_frame.configure(bg = "thistle1")
        btn_frame.place(relx = 0, rely = 0.93, relwidth = 1, relheight = 0.07)

        b_nc_back = Button(btn_frame, text = "Back", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = back2home)
        b_nc_back.grid(row = 0, column = 0, padx = 2, pady = 0)

        b_nc_save = Button(btn_frame, text = "Save", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = newcontactdb)
        b_nc_save.grid(row = 0, column = 1, padx = 0, pady = 0)

# **************************************** This is Back Button on Main Contact List ****************************************

    def back():
        win.destroy()      
    
    btn_frame = Frame(win)
    btn_frame.configure(bg = "thistle1")
    btn_frame.place(relx = 0, rely = 0.935, relwidth = 1, relheight = 0.07)

    b_hs_back = Button(btn_frame, text = "Back", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = back)
    b_hs_back.grid(row = 0, column = 0, padx = 2, pady = 0)

    b_hs_new = Button(btn_frame, text = "New Contact", width = 20, font = ("Arial", 16, "bold"), bg = "purple", fg = "white", bd = 2, command = newcontact)
    b_hs_new.grid(row = 0, column = 1, padx = 0, pady = 0)

    contactlist()


homescreen()
win.mainloop()
```
