```python
import tkinter as tk
from tkinter import ttk
from tkinter.ttk import Combobox
from tkinter import messagebox
import sqlite3 as sl

colors = ['wheat1', 'brown', 'white', 'black']
ff = ['Arial', 'calibri']
labels = ['Name', 'Mobile No.', 'Telephone No.', 'Email-ID', 'Address', 'Relation']
rel = ['Mother', 'Father', 'Sister', 'Brother', 'Friend', 'Cousin']
data = []
m_frame = None
btn_frame = None

#   database creation
try:
    con = sl.connect(database= "contactbookdb.sqlite3")
    cur = con.cursor()
    cur.execute("""CREATE TABLE contactbook(
                cb INTEGER PRIMARY KEY AUTOINCREMENT, 
                img TEXT,
                name TEXT,
                mob INTEGER,
                telephone INTEGER,
                email TEXT,
                address TEXT,
                relation TEXT
                )""")
    con.commit()
    print("Table created successfully.")
except:
    print("Something went wrong! Might be table already created.")
con.close()


def getallcb():
    global data
    #   database extraction to show all contacts
    conobj = sl.connect(database= "contactbookdb.sqlite3")
    cur = conobj.cursor()
    cur.execute("select * from contactbook")
    data = cur.fetchall()
    conobj.close()

#   ***************************************** This is Search Function ********************************************************

def search():
    global data
    s = e_search.get()
    if s != "":
        # database search for text entered 
        conobj = sl.connect(database= "contactbookdb.sqlite3")
        cur = conobj.cursor()
        cur.execute(f"SELECT * FROM contactbook WHERE name LIKE \'%{s}%\'")
        data = cur.fetchall()
        conobj.close()
        homescreen()
    else:
        getallcb()
        homescreen()


win = tk.Tk()
win.geometry("550x700")
win.resizable(width = False, height = False)
win.configure(bg = colors[0])
win.title("Contact Book")
icon = tk.PhotoImage(file = "logo.png")
win.iconphoto(False, icon)

h_frame = tk.Frame(win)
h_frame.configure(bg = colors[2])
h_frame.place(relx = 0, rely = 0, relwidth = 1, relheight = 0.075)

l_head = tk.Label(h_frame, text = "Contact Book", image = icon, font = (ff[0], 24, "bold", "underline"), bg = colors[2], fg = colors[1], width = 50)
l_head.grid(row = 0, column = 0, padx = 5)

e_search = tk.Entry(h_frame, font = (ff[0], 16, "bold"), bg = colors[2], fg = colors[1], width = 32, bd = 1)
e_search.grid(row = 0, column = 1, padx = 8, pady = (10, 0))

b_search = tk.Button(h_frame, text = "Search", font = (ff[0], 10, "bold"), bg = colors[1], fg = colors[2], command = search)
b_search.grid(row = 0, column = 2, padx = (10, 2), pady = (10, 2))


def newcontact():
    global m_frame, btn_frame
    if m_frame is not None and btn_frame is not None: 
        m_frame.destroy()
        btn_frame.destroy()

    m_frame = tk.Frame(win)
    m_frame.configure(bg = colors[0])
    m_frame.place(relx = 0, rely = 0.076, relwidth = 1, relheight = 0.9)

    nc_frame = tk.Frame(m_frame)
    nc_frame.configure(bg = colors[0])
    nc_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)
       
    l_nc_name = tk.Label(nc_frame, text = "Name:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_name.grid(row = 1, column = 0, pady = 10)

    e_nc_name = tk.Entry(nc_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_nc_name.grid(row = 1, column = 2, pady = 10)

    l_nc_mob = tk.Label(nc_frame, text = "Mobile No.:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_mob.grid(row = 2, column = 0, pady = 10)

    e_nc_mob = tk.Entry(nc_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_nc_mob.grid(row = 2, column = 2, pady = 10)

    l_nc_tele = tk.Label(nc_frame, text = "Telephone No.:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_tele.grid(row = 3, column = 0, pady = 10)

    e_nc_tele = tk.Entry(nc_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_nc_tele.grid(row = 3, column = 2, pady = 10)

    l_nc_email = tk.Label(nc_frame, text = "Email ID:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_email.grid(row = 4, column = 0, pady = 10)         
        
    e_nc_email = tk.Entry(nc_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_nc_email.grid(row = 4, column = 2, pady = 10)

    l_nc_addr = tk.Label(nc_frame, text = "Address:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_addr.grid(row = 5, column = 0, pady = 10)

    e_nc_addr = tk.Entry(nc_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_nc_addr.grid(row = 5, column = 2, pady = 10)

    l_nc_rel = tk.Label(nc_frame, text = "Relation:", fg = colors[3], font = (ff[0], 16, ""), bg = colors[0])
    l_nc_rel.grid(row = 6, column = 0, pady = 10)

    e_nc_rel = Combobox(nc_frame, width = 23, font = (ff[1], 16, "bold"), values = rel)
    e_nc_rel.current(0)
    e_nc_rel.grid(row = 6, column = 2, pady = 10)

    def newcontactdb():
        name = e_nc_name.get()
        email = e_nc_email.get().lower()
        mob = e_nc_mob.get()
        tele = e_nc_tele.get()
        addr = e_nc_addr.get().title()
        rel = e_nc_rel.get()

        if name == "" and mob == "":
            messagebox.showwarning(title = "Warning", message = "Invalid input. Enter the valid Details!")
            homescreen()
        else: 
            pass;

        #   database insertion
        conobj = sl.connect(database= "contactbookdb.sqlite3")
        cur = conobj.cursor()
        cur.execute("INSERT INTO contactbook(name, mob, telephone, email, address, relation) VALUES(?, ?, ?, ?, ?, ?)", (name, mob, tele, email, addr, rel))
        conobj.commit()
        conobj.close()
        messagebox.showinfo("Message", "New contact added successfully!")
        m_frame.destroy()
        getallcb()
        homescreen()

    def back2home():
        m_frame.destroy()
        btn_frame.destroy()
        getallcb()
        homescreen()

    btn_frame = tk.Frame(win)
    btn_frame.configure(bg = colors[0])
    btn_frame.place(relx = 0, rely = 0.935, relwidth = 1, relheight = 0.07)

    b_nc_back = tk.Button(btn_frame, text = "Back", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = back2home)
    b_nc_back.grid(row = 0, column = 0, padx = 2, pady = 0)

    b_nc_save = tk.Button(btn_frame, text = "Save", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = newcontactdb)
    b_nc_save.grid(row = 0, column = 1, padx = 0, pady = 0)


# ************************************************ This Edit/Update Contact Details ***************************************************

def editcontact(cb_id, cd):
    global m_frame, btn_frame
    if m_frame is not None and btn_frame is not None: 
        m_frame.destroy()
        btn_frame.destroy()
    
    m_frame=tk.Frame(win)
    m_frame.configure(bg = colors[0])
    m_frame.place(relx = 0, rely = 0.076, relwidth = 1, relheight = 0.9)

    edit_frame=tk.Frame(m_frame)
    edit_frame.configure(bg = colors[0])
    edit_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)

    def back2cd():
        m_frame.destroy()
        btn_frame.destroy()
        contactdetails(cb_id)

    def updatedb():
        n = e_edit_name.get()
        m = e_edit_mob.get()
        t = e_edit_tele.get()
        e = e_edit_email.get()
        a = e_edit_addr.get()
        r = e_edit_rel.get()

        #   database updation
        conobj = sl.connect(database = "contactbookdb.sqlite3")
        cur = conobj.cursor()
        cur.execute("UPDATE contactbook SET name = ?, mob = ?, telephone = ?, email = ?, address = ?, relation = ? WHERE cb = ?", (n, m, t, e, a, r, cb_id))
        conobj.commit()
        conobj.close()
        messagebox.showinfo("Message", "Contact Updated successfully!")
        m_frame.destroy()
        contactdetails(cb_id)

    for i in range(len(labels)):
        l_edit_name = tk.Label(edit_frame, text = labels[i], fg = colors[3], font = (ff[0], 16, "normal"), bg = colors[0])
        l_edit_name.grid(row = i+1, column = 0, pady = 10)

    e_edit_name = tk.Entry(edit_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_edit_name.grid(row = 1, column = 2, pady = 10)

    e_edit_mob = tk.Entry(edit_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_edit_mob.grid(row = 2, column = 2, pady = 10)                

    e_edit_tele = tk.Entry(edit_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_edit_tele.grid(row = 3, column = 2, pady = 10)              

    e_edit_email = tk.Entry(edit_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_edit_email.grid(row = 4, column = 2, pady = 10)

    e_edit_addr = tk.Entry(edit_frame, width = 25, font = (ff[1], 16, "bold"), fg = colors[1], bd = 3)
    e_edit_addr.grid(row = 5, column = 2, pady = 10)

    e_edit_rel = Combobox(edit_frame, width = 25, font = (ff[1], 14, "bold"), values = rel)
    e_edit_rel.grid(row = 6, column = 2, pady = 10)

    e_edit_name.insert(0, cd[2])
    e_edit_mob.insert(0, cd[3])
    e_edit_tele.insert(0, cd[4])
    e_edit_email.insert(0, cd[5])
    e_edit_addr.insert(0, cd[6])
    r =rel.index(cd[7])
    e_edit_rel.current(r)

    btn_frame = tk.Frame(win)
    btn_frame.configure(bg = colors[0])
    btn_frame.place(relx = 0, rely = 0.935, relwidth = 1, relheight = 0.07)

    b_edit_back = tk.Button(btn_frame, text = "Back", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = back2cd)
    b_edit_back.grid(row = 0, column = 0, padx = 1.5, pady = 0)

    b_edit_save = tk.Button(btn_frame, text = "Save", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = updatedb)
    b_edit_save.grid(row = 0, column = 1, padx = 1.5, pady = 0)


# ***************************************** This is Contact Details Display Screen *****************************************************

def contactdetails(cb_id):
    global m_frame, btn_frame
    if m_frame is not None and btn_frame is not None: 
        m_frame.destroy()
        btn_frame.destroy()

    m_frame = tk.Frame(win)
    m_frame.configure(bg = colors[0])
    m_frame.place(relx = 0, rely = 0.076, relwidth = 1, relheight = 0.9)

    cd_frame = tk.Frame(m_frame)
    cd_frame.configure(bg = colors[0])
    cd_frame.place(relx = 0.1, rely = 0.15, relheight = 0.75, relwidth  = 0.8)

    # database extraction
    conobj = sl.connect(database= "contactbookdb.sqlite3")
    cur = conobj.cursor()
    cur.execute("SELECT * FROM contactbook  WHERE cb = ?", (str(cb_id)))
    cd = cur.fetchone()
    conobj.close()

    #   contact details display
    for j in range(len(labels)):
        cd_label = tk.Label(cd_frame, text = labels[j], width = 12, fg = colors[3], font = (ff[0], 16, ""), bg = colors[0], anchor ='w')
        cd_label.grid(row = j, column = 0, pady = 10)

        cd_details = tk.Label(cd_frame, text = cd[j+2], width = 26, wraplength = 275, font = (ff[1], 14, ""), fg = colors[1], bd = 3, bg = colors[0], anchor = "w")
        cd_details.grid(row = j, column = 2, pady = 10)

    def back2list():
        m_frame.destroy()
        btn_frame.destroy()
        getallcb()
        homescreen()

    def deletedb():
        confrm = messagebox.askokcancel(message = f"Do you really want to Delete contact {cd[2]}.")
        
        if confrm:
            conobj = sl.connect(database= "contactbookdb.sqlite3")
            cur = conobj.cursor()
            cur.execute("DELETE FROM contactbook WHERE cb = ?", (str(cb_id)))
            conobj.commit()
            conobj.close()
                    
            m_frame.destroy()
            getallcb()
            homescreen()
        else:
            pass;

    btn_frame = tk.Frame(win)
    btn_frame.configure(bg = colors[0])
    btn_frame.place(relx = 0, rely = 0.9375, relwidth = 1, relheight = 0.07)

    b_cd_back = tk.Button(btn_frame, text = "Delete", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = deletedb)
    b_cd_back.grid(row = 0, column = 0, padx = 1.5, pady = 0)

    b_cd_save = tk.Button(btn_frame, text = "Edit", width = 20, command = lambda : editcontact(cb_id, cd), font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2)
    b_cd_save.grid(row = 0, column = 1, padx = 1.5, pady = 0)

    b_cd_back = tk.Button(m_frame, text = "Back", width = 5, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = back2list)
    b_cd_back.place(relx = 0, rely = 0)


# ***************************************** This is Contact List Function of Home Screen ************************************************

def homescreen():
    global m_frame, btn_frame
    if m_frame is not None and btn_frame is not None: 
        m_frame.destroy()
        btn_frame.destroy()
    
    m_frame = tk.Frame(win)
    m_frame.configure(bg = colors[0])
    m_frame.place(relx = 0, rely = 0.076, relwidth = 1, relheight = 0.9)
        
    canvas = tk.Canvas(m_frame)
    canvas.configure(bg = colors[0], bd = 0)
    canvas.pack(side = tk.LEFT, fill = 'both', expand = True)
        
    cl_scrollbar = ttk.Scrollbar(m_frame, orient = "vertical", command = canvas.yview)
    cl_scrollbar.pack(side = tk.RIGHT, fill = "y")

    canvas.configure(yscrollcommand = cl_scrollbar.set)
    canvas.bind('<Configure>', lambda e: canvas.configure(scrollregion=canvas.bbox("all")))

    cl_frame = tk.Frame(canvas)
    cl_frame.configure(bg = colors[0])
    canvas.create_window((75, 0), window = cl_frame, anchor = "nw")

    global data
    slno = []
    for j in range(len(data)): slno.append(data[j][0])

    n = 0 
    for i in slno:
        c = data[n]

        profile_pic = tk.PhotoImage(file = "human-icon.png")
        b_cl_name = tk.Button(cl_frame, text = c[2], image = profile_pic, width = 350, compound = "left", command = lambda i = i : contactdetails(i),
                            bd = 0, font = (ff[1], 18, ""), bg = colors[0], fg = colors[1], anchor = 'w', padx = 20)
        b_cl_name.image = profile_pic
        b_cl_name.grid(row = n, column = 0, pady = 10)
        n+=1

    # **************************************** This is Back Button on Main Contact List ****************************************

    def back():
        win.destroy()      

    btn_frame = tk.Frame(win)
    btn_frame.configure(bg = colors[0])
    btn_frame.place(relx = 0, rely = 0.935, relwidth = 1, relheight = 0.07)

    b_hs_back = tk.Button(btn_frame, text = "Back", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = back)
    b_hs_back.grid(row = 0, column = 0, padx = 2, pady = 0)

    b_hs_new = tk.Button(btn_frame, text = "New Contact", width = 20, font = (ff[0], 16, "bold"), bg = colors[1], fg = colors[2], bd = 2, command = newcontact)
    b_hs_new.grid(row = 0, column = 1, padx = 0, pady = 0)


getallcb()
homescreen()

win.mainloop()
```
